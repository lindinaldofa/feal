#Formação Econômica da América Latina

##Apresentação - Agricultura Tropical

###Fonte dos dados: BÉRTOLA, Luís; OCAMPO, José Antonio. O desenvolvimento econômico da América Latina desde a independência. Rio de Janeiro: Alta Books, 2019.

####GK$: dólares internacionais Geary-Khamys de 1990

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
import plotly.graph_objs as go
import plotly.express as px

init_notebook_mode(connected=True)
cf.go_offline()
```

```python
dataPIB = pd.read_csv('PIBnom.csv', delimiter=';', index_col='Ano')
```

```python
figvot = px.line(dataPIB, height = 400, labels={'value':'GK$','variable':'Países'})
figvot.update_layout(font_size=16,  title="PIB dos países latino-americanos de agricultura tropical", 
        title_font_size = 24)
figvot.show()
```

```python
dataPIBp = pd.read_csv('PIBpercap.csv', delimiter=';', index_col='Ano')
```

```python
figvot = px.line(dataPIBp, height = 400, labels={'value':'GK$','variable':'Países'})
figvot.update_layout(font_size=16,  title="PIB per capita dos países latino-americanos de agricultura tropical", 
        title_font_size = 24)
figvot.show()
```

#Comparação em relação ao Brasil

```python
from ipywidgets import interact
print('Selecione o país que deseja comparar.')
@interact(País=['Colombia','Cuba','Republica dominicana','Venezuela','Panama','Costa Rica'])

def iplot(País):
    pais = País
    dataPIB[['Brasil', pais]].iplot(kind='spread', title='Comparação do PIB em relação ao Brasil')
    dataPIBp[['Brasil', pais]].iplot(kind='spread', title='Comparação do PIB per capita em relação ao Brasil')
```
